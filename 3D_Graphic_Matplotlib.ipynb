# 3D Graphic Example

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos de ejemplo
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
z = np.array([6, 7, 8, 9, 10])

# Ajustar una regresión lineal en 3D
coefficients = np.polyfit(x, y, 1)
slope, intercept = coefficients

# Generar puntos para la línea de regresión en 3D
x_future = np.linspace(min(x), max(x), 100)
y_future = np.linspace(min(y), max(y), 100)
x_future, y_future = np.meshgrid(x_future, y_future)
z_intercept = slope * x_future + intercept

# Crear la figura y el espacio tridimensional
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar los puntos de datos
ax.scatter(x, y, z, c='red', label='Datos')

# Graficar la línea de regresión
ax.plot_surface(x_future, y_future, z_intercept, alpha=0.5, label='Regresión')

# Configurar los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar el gráfico
plt.show()
```
