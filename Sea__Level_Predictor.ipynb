# Sea Level Predictor Using Python


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
# Read data from file
df = pd.read_csv(r"/content/sample_data/epa-sea-level.csv")

# Create scatter plot
plt.scatter(df["Year"], df["CSIRO Adjusted Sea Level"])
    

# Create first line of best fit
slope, intercept, rvalue, pvalue, stderr = linregress(df["Year"], df["CSIRO Adjusted Sea Level"])
x_future = np.arange(1880, 2050)
y_predicted = slope * x_future + intercept
plt.plot(x_future, y_predicted, color='red', label='line of best fit')

# Create second line of best fit
slope1, intercept1, rvalue1, pvalue1, stderr1 = linregress(df["Year"][df["Year"] >= 2000], df["CSIRO Adjusted Sea Level"][df["Year"] >= 2000])
x_future1 = np.arange(2000, 2050)
y_predicted1 = slope1 * x_future1 + intercept1
plt.plot(x_future1, y_predicted1, color='green', label='line of best fit begin at 2000')

# Add labels and title
plt.xlabel("Years")
plt.ylabel("Sea Level (inches)")
plt.title("Rise in Sea Level")

plt.show()
```
